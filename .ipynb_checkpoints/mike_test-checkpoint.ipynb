```python
import pandas as pd
from sklearn import svm
```

```python
#get data
data = pd.read_csv('FY2019_070_Contracts_Full_20200110_1.csv')
print(data.shape)
```

```python
#create binary variable
def over_thresh(fao):
    if fao < -1000:
        return 1
    else:
        return 0
    

data['significant_deobligation'] = data.apply(lambda x : over_thresh(x['federal_action_obligation']), axis = 1)

data.head(5)
```

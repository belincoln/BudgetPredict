```python
import glob

# Create a list of the file names we want to import (This will be all csv files inside the data folder)
list_of_csvs = glob.glob("/*.csv")

# Create an empty list to store dfs as we read them in
li = []

for csv in list_of_csvs:
    df = pd.read_csv(csv,header = 0)
    li.append(df)

df = pd.concat(li, axis = 0, ignore_index = True)
```
